```python
!pip install opencv-python dlib

import cv2
import dlib
import time
from google.colab.patches import cv2_imshow
from google.colab import files
from skimage import io

!wget http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2
!bzip2 -d shape_predictor_68_face_landmarks.dat.bz2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

detector = dlib.get_frontal_face_detector()

predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

def detect_faces_image_opencv(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Erro ao carregar a imagem: {image_path}")
        return
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    cv2_imshow(img)

# Função para detectar faces em uma imagem estática com Dlib
def detect_faces_image_dlib(image_path):
    img = io.imread(image_path)
    faces = detector(img, 1)
    for face in faces:
        x1 = face.left()
        y1 = face.top()
        x2 = face.right()
        y2 = face.bottom()
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
    io.imshow(img)

def detect_landmarks_dlib(image_path):
    img = io.imread(image_path)
    faces = detector(img, 1)
    for face in faces:
        landmarks = predictor(img, face)
        for n in range(0, 68):
            x = landmarks.part(n).x
            y = landmarks.part(n).y
            cv2.circle(img, (x, y), 2, (0, 255, 0), -1)
    io.imshow(img)

def compare_face_detection(image_path):
    print("Comparação entre OpenCV e Dlib")

    # OpenCV
    start_time = time.time()
    detect_faces_image_opencv(image_path)
    end_time = time.time()
    print(f"Tempo de execução com OpenCV: {end_time - start_time} segundos")

    # Dlib
    start_time = time.time()
    detect_faces_image_dlib(image_path)
    end_time = time.time()
    print(f"Tempo de execução com Dlib: {end_time - start_time} segundos")

def main():
    print("Por favor, faça o upload de uma imagem para realizar a detecção de faces.")
    uploaded = files.upload()
    image_path = list(uploaded.keys())[0]

    print("Detecção de faces com OpenCV:")
    detect_faces_image_opencv(image_path)

    print("Detecção de faces com Dlib:")
    detect_faces_image_dlib(image_path)

    print("Comparação de detecção de faces entre OpenCV e Dlib:")
    compare_face_detection(image_path)

    print("Detecção de landmarks faciais com Dlib:")
    detect_landmarks_dlib(image_path)

main()
```
